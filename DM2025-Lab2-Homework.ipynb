## 1. Preprocessing Steps

### import data

```python
import pandas as pd
import json
with open('./kaggle/final_posts.json','r',encoding='utf-8') as file:
  post=json.load(file)
```

```python
row=[]
for data in post:
  p=data["root"]["_source"]["post"]
  row.append({
      "id":p["post_id"],
      "text":p["text"],
      "hashtags":p["hashtags"],

  })
df_post=pd.DataFrame(row)
df_post
```

```python
df_emotion=pd.read_csv("./kaggle/emotion.csv")
df_emotion
```

```python
df_identification=pd.read_csv("./kaggle/data_identification.csv")
df_identification
```

### Put hashtag into text
Since there is lots of empty in the hashtag column so we add hashtag coulum into text column.
Then, with input of hashtag column, when we predict emotion of text, might be more accurate.

```python
df_post['hashtags']=df_post['hashtags'].fillna('')  
df_post['hashtag_text']=df_post['hashtags'].apply(
    lambda tags: ' '.join([tag.replace('#','') for tag in tags if len(tag)>1]) 
)
df_post['text']=df_post['text']+' '+df_post['hashtag_text']
```

### split the data into training and testing data
by data_identification.csv

```python
df_identification=pd.read_csv("./kaggle/data_identification.csv")
df_identification

df=pd.merge(df_post,df_emotion,on="id",how="left")
df
df_all=pd.merge(df,df_identification,on="id",how="left")
df_train=df_all[df_all["split"]=="train"][['id','text','emotion']]
df_test=df_all[df_all["split"]=="test"][['id','text']]

```

```python
df_test
```

```python
#check the distribution of emotions in the training data
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


labels=df_train['emotion'].unique()
post_total=len(df_train)
df1=df_train.groupby(['emotion']).count()['text']
df1=df1.apply(lambda x: round(x*100/post_total,3))

fig, ax=plt.subplots(figsize=(5,3))
plt.bar(df1.index,df1.values)

plt.ylabel('% of instances')
plt.xlabel('Emotion')
plt.title('Emotion distribution')
plt.grid(True)
plt.show()
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
nltk.download('punkt_tab')
```

```python
bow_vectorizer=CountVectorizer(max_features=500, tokenizer=nltk.word_tokenize)
bow_vectorizer.fit(df_train['text'])

df_train_bow=bow_vectorizer.transform(df_train['text'])
df_test_bow=bow_vectorizer.transform(df_test['text'])
```

## 3. Model Implementation Steps

```python
### Add the code related to the model implementation steps in cells inside this section
```

### 3.1 Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

X_train=df_train_bow
y_train=df_train['emotion']
X_test=df_test_bow

dt_model=DecisionTreeClassifier(random_state=1)
dt_model=dt_model.fit(X_train,y_train)

y_test_pred=dt_model.predict(X_test)

```

```python
df_test['emotion_pred']=y_test_pred
df_test
```

```python
submission=pd.read_csv("./kaggle/samplesubmission.csv")
predicted_emotions=df_test[['id','emotion_pred']]
submission=pd.merge(submission[['id']],predicted_emotions,on='id',how='left')
submission.rename(columns={'emotion_pred':'emotion'}, inplace=True)
submission.to_csv("submission.csv",index=False)
```

### 3.2 Deep Learning

```python
import keras
from sklearn.preprocessing import LabelEncoder

labelencoder=LabelEncoder()
labelencoder.fit(df_train['emotion'])
```

```python
X_train=df_train_bow
y_train=df_train['emotion']
X_test=df_test_bow
```

```python
y_train=keras.utils.to_categorical(labelencoder.transform(df_train['emotion']))
X_train_dense=X_train.toarray()
X_test_dense=X_test.toarray()
```

```python
from keras.models import Model
from keras.layers import Input, Dense
from keras.layers import ReLU, Softmax

input_shape=X_train.shape[1]
output_shape=len(labelencoder.classes_)


model_input=Input(shape=(input_shape,))

x=Dense(128,activation='relu')(model_input)
x=Dense(128,activation='relu')(x)
model_output=Dense(output_shape, activation='softmax')(x)

model=Model(inputs=model_input, outputs=model_output)
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from keras.callbacks import CSVLogger

csv_logger=CSVLogger('training_log.csv')


epochs=25
batch_size=32

history=model.fit(
    X_train_dense, y_train,
    epochs=epochs,
    batch_size=batch_size,
    callbacks=[csv_logger]
)
```

```python
pred_result=model.predict(X_test_dense, batch_size=128)
y_pred_label=np.argmax(pred_result ,axis=1)
emotion_pred_dl=labelencoder.inverse_transform(y_pred_label)
df_test["dl_pred"]=emotion_pred_dl
df_test
```

```python
submission_2=pd.read_csv("./kaggle/samplesubmission.csv")
submission_2
predicted_emotions_dl=df_test[['id','dl_pred']]
submission_2=pd.merge(submission_2[['id']],predicted_emotions_dl,on='id',how='left')
submission_2.rename(columns={'emotion_pred':'emotion'}, inplace=True)
submission_2.to_csv("submission_2.csv",index=False)
```
